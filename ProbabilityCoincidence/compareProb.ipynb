```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import array
import sys
sys.path.insert(0, '/Users/sdporzio/HighVoltageTask')
import HvPackages.probability as prob
```

```python
nMonths = 3
tWindow = 30
nHV = 100
nPMT = 100

x = array.array("f",[])
y1 = array.array("f",[])
y2 = array.array("f",[])
y3 = array.array("f",[])

i = 0
for var in np.arange(0.1,5,0.1):
    nMonths = var
    x.append(var)
    y1.append(prob.prob1(nMonths,tWindow,nHV,nPMT))
    y2.append(prob.prob2(nMonths,tWindow,nHV,nPMT))
    y3.append(prob.prob3(nMonths,tWindow,nHV,nPMT))
    i+=1
    
plt.figure(figsize=(15,5), dpi=80)
plt.plot(x, y1, color="red", linewidth=3, linestyle="-", label="p1")
plt.plot(x, y2, color="blue", linewidth=3, linestyle=":", label="p2")
plt.plot(x, y3, color="green", linewidth=3, linestyle="--", label="p3")
plt.legend(loc='lower right', frameon=True)
plt.title('Number of Months vs. Probability')
plt.xlabel('# Months')
plt.ylabel('Probability')
plt.show()
```
