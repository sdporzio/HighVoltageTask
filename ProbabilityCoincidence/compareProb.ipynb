```python
import ROOT
ROOT.gROOT.SetBatch(1)
def destruct(object_):
    if issubclass(type(object_), ROOT.TObject):
        object_.IsA().Destructor(object_)
from IPython.display import Image as Plot
g_xres = 950
g_yres = 425


import numpy as np
import sys
sys.path.insert(0, '/Users/sdporzio/HighVoltageTask')
import HvPackages.probability as prob
```

```python
nMonths = 3
tWindow = 30
nHV = 100
nPMT = 100

c1 = ROOT.TCanvas("canvas1","",g_xres,g_yres)
g1 = ROOT.TGraph()
g2 = ROOT.TGraph()
g3 = ROOT.TGraph()
g1.SetLineColor(ROOT.kBlue)
g2.SetLineColor(ROOT.kRed)
g3.SetLineColor(ROOT.kGreen)
g1.SetTitle("Probability vs. Total Time;# Months;Probability")

i = 0
for var in np.arange(0.1,5,0.1):
    nMonths = var
    g1.SetPoint(i,var,prob.prob1(nMonths,tWindow,nHV,nPMT))
    g2.SetPoint(i,var,prob.prob2(nMonths,tWindow,nHV,nPMT))
    g3.SetPoint(i,var,prob.prob3(nMonths,tWindow,nHV,nPMT))
    i+=1
    
g1.Draw("ALP")
g2.Draw("SAME")
g3.Draw("SAME")

c1.SaveAs("Png/histo1.png")
destruct(c1)
destruct(g1)
destruct(g2)
destruct(g3)
Plot("Png/histo1.png")
```
